```python
from specpol.ritz import multiplication_operator_ritz

import matplotlib.pyplot as plt
```

```python
def step_slope(x):
    return x if x < 1/2 else x + 1/2

spec_step_slope = multiplication_operator_ritz(step_slope, 100, 321)
plt.scatter(spec_step_slope.real, spec_step_slope.imag)
```

```python
import numpy as np
def circle(x):
    return np.exp(1j*2*np.pi*x)

circle_spec = multiplication_operator_ritz(circle, 250, 321)
plt.scatter(circle_spec.real, circle_spec.imag)
```
