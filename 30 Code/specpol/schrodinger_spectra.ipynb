```python
import numpy as np
import matplotlib.pyplot as plt

from typing import Callable
```

```python
laplacian = lambda n: np.diag(np.ones(n), k=1) + np.diag(np.ones(n), k=-1)
```

```python
def schrodinger(n: int, potential: Callable) -> np.array:
    return laplacian(n) + np.diag([potential(k) for k in range(n+1)])
```

```python
eigvals = np.linalg.eigvalsh(schrodinger(5000, lambda n: n**(1/2)))
```

```python
plt.scatter(eigvals, [0]*len(eigvals))
```
